Below we go through several types of charts you can plot with python

Bar chart can be used to make comparisons between several categories.  For example below we compare the points scored by three basketball players

```python
%matplotlib inline
import matplotlib.pyplot as plt

print (type(plt))

lebron=27
durant=24
grant=26

names=['lebron','durant','grant']
```

```python
plt.bar(1, lebron, color='blue')
plt.bar(2, durant, color='brown')
plt.bar(3, grant, color='orange')
plt.legend(names)
plt.title("Basketball players")
plt.ylabel("Points")

# note a little bit of obtuse code is needed to get rid of the x-ticks... 
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)

plt.show()
```

Scatterplot is a way to show a relationship between an independent variable (on the x axis) and a dependent variable (on the y axis).  In the below case we show hypothetical point vs age for basketball players. 

```python
age = [21, 33, 40]
points = [12, 28.3, 11]
plt.scatter(age, points)
plt.xlabel('age')
plt.ylabel('points')
plt.title('points vs age')
```

Bubbleplot is a way to show the relationship between an independent variable and two dependent variables, where the size of the bubble represents the second independent variable.  For example we could show points and assista vs age

```python
assists = [1000,2000, 3000]
plt.scatter(age, points, assists)
plt.xlabel('age')
plt.ylabel('points')
plt.title('points vs age (size of bubble is assists)')
```

Pie chart is yet another way to compare categories.  Pie charts are useful when we are comparing items that are a share of a larger total.  For example comparing points scored by Players on the same team.  Note in this case we initialize the figure before calling 'pie'.  That step is optional but it is done so that we can set the size of the figure.

```python
bucks_points=[29.7, 21.1, 15.8, 10.8, 9.6]
bucks_names=['Antetokounmpo', 'Middleton', 'Bledsoe', 'Lopez', 'Hill']
plt.figure(figsize=(20,10))
temp=plt.pie(bucks_points, labels=bucks_names)
```
